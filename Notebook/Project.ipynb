Around 12 November 2025, the company reorganized its business operations by separating its commercial vehicle division into a distinct listed entity, represented by the ticker TMCV.NS. Prior to this date, all stock market data was recorded under the unified ticker TATAMOTORS.NS, which reflected the combined performance of both passenger and commercial vehicle segments.

Following the demerger, financial data platforms such as Yahoo Finance began treating the newly formed entity as a separate financial instrument. As a result, historical data before 11 November 2025 remains associated with the original combined company, while data from 12 November 2025 onward is recorded under the new ticker. This creates a clear discontinuity in the dataset, and APIs like yfinance are unable to fetch the entire dataset using a single ticker symbol.

This issue is not a technical error but a structural break in the company’s financial identity. The pre-demerger data represents a diversified automotive business, whereas the post-demerger data reflects a more specialized segment. Due to this difference in business structure, revenue composition, and financial behavior, combining both datasets directly may lead to misleading trends, incorrect model training, and inaccurate forecasting results.

STEP 1 — Load & Reshape Your NIFTY 50 Data

```python
import pandas as pd

# Load data
data = pd.read_csv("nifty50_5y_data.csv", header=[0, 1], index_col=0)

# Fix index (correct format)
data.index = pd.to_datetime(data.index, dayfirst=True)

# Convert to long format
df = data.stack(level=0, future_stack=True).reset_index()

# Rename columns
df.columns = ["Date", "Ticker", "Open", "High", "Low", "Close", "Volume"]

print(df.head())
```

STEP 2 — Clean Data

```python
df = df.dropna()
df = df.sort_values(["Ticker", "Date"])
df.reset_index(drop=True, inplace=True)
```

STEP 3 — Feature Engineering (IMPORTANT)
RSI, Moving Average, Returns

```python
import numpy as np

def compute_rsi(series, period=14):
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Apply per stock
df["Return"] = df.groupby("Ticker")["Close"].pct_change()

df["MA20"] = df.groupby("Ticker")["Close"].transform(lambda x: x.rolling(20).mean())
df["MA50"] = df.groupby("Ticker")["Close"].transform(lambda x: x.rolling(50).mean())

df["RSI"] = df.groupby("Ticker")["Close"].transform(lambda x: compute_rsi(x))
```

STEP 4 — Target Variable (Prediction Goal)

```python
## Predict 5-day return

df["Target"] = df.groupby("Ticker")["Close"].shift(-5)
df["Target_Return"] = (df["Target"] - df["Close"]) / df["Close"]
```

STEP 5 — Drop NaN after feature creation

```python
df = df.dropna()
```

STEP 6 — Train Machine Learning Model

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

features = ["Return", "MA20", "MA50", "RSI"]
X = df[features]
y = df["Target_Return"]

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.2)

model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, y_train)

preds = model.predict(X_test)
```

STEP 7 — Evaluate Model

```python
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE:", rmse)
```

```python
import pandas as pd
import numpy as np

# Load data
data = pd.read_csv("nifty50_5y_data.csv", header=[0,1], index_col=0)
data.index = pd.to_datetime(data.index)

# Extract tickers
tickers = data.columns.get_level_values(0).unique()

print("Total Stocks:", len(tickers))
```

STEP 2 — Feature Engineering (Indicators)

```python
def add_indicators(df):
    df = df.copy()

    # Moving Average
    df['MA20'] = df['Close'].rolling(20).mean()

    # RSI
    delta = df['Close'].diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)

    avg_gain = gain.rolling(14).mean()
    avg_loss = loss.rolling(14).mean()

    rs = avg_gain / avg_loss
    df['RSI'] = 100 - (100 / (1 + rs))

    # MACD
    exp1 = df['Close'].ewm(span=12).mean()
    exp2 = df['Close'].ewm(span=26).mean()
    df['MACD'] = exp1 - exp2

    return df.dropna()
```

STEP 3 — Prepare Dataset

```python
all_data = {}

for ticker in tickers:
    df = data[ticker]
    df = add_indicators(df)

    # Target (next day price movement)
    df['Target'] = (df['Close'].shift(-1) > df['Close']).astype(int)

    df = df.dropna()
    all_data[ticker] = df
```

STEP 4 — Train Multiple ML Models (6 Models)

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier

models = {
    "Logistic": LogisticRegression(),
    "RandomForest": RandomForestClassifier(),
    "SVM": SVC(),
    "DecisionTree": DecisionTreeClassifier(),
    "GradientBoost": GradientBoostingClassifier(),
    "XGBoost": XGBClassifier()
}
```

STEP 5 — Train + Predict for Each Stock

```python
results = []

for ticker, df in all_data.items():

    X = df[['Close', 'MA20', 'RSI', 'MACD']]
    y = df['Target']

    X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False)

    predictions = []

    for name, model in models.items():
        model.fit(X_train, y_train)
        pred = model.predict(X_test)

        acc = accuracy_score(y_test, pred)
        predictions.append(pred[-1])  # last prediction

    # Majority Voting
    final_pred = round(np.mean(predictions))

    latest = df.iloc[-1]

    # Recommendation Logic
    if final_pred == 1 and latest['RSI'] < 70:
        rec = "BUY"
    elif final_pred == 0 and latest['RSI'] > 30:
        rec = "SELL"
    else:
        rec = "HOLD"

    results.append({
        "Stock": ticker,
        "Prediction": final_pred,
        "RSI": latest['RSI'],
        "MACD": latest['MACD'],
        "Recommendation": rec
    })

results_df = pd.DataFrame(results)
print(results_df.head())
```

STEP-BY-STEP COMPLETE CODE (MODEL-BASED CHART)

1. Prepare data (features + target)

```python
import pandas as pd
import numpy as np

def prepare_data(df):
    df = df.copy()

    df["Return"] = df["Close"].pct_change()

    df["MA20"] = df["Close"].rolling(20).mean()
    df["MA50"] = df["Close"].rolling(50).mean()

    df["Volatility"] = df["Return"].rolling(10).std()

    # Target: next day up/down
    df["Target"] = (df["Return"].shift(-1) > 0).astype(int)

    df = df.dropna()

    return df
```

2. Train ALL models

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

def train_models(df, models):
    X = df[["MA20", "MA50", "Volatility"]]
    y = df["Target"]

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, shuffle=False
    )

    trained_models = {}
    results = {}

    for name, model in models.items():
        model.fit(X_train, y_train)
        preds = model.predict(X_test)

        acc = accuracy_score(y_test, preds)

        trained_models[name] = model
        results[name] = acc

        print(f"{name}: {acc:.2f}")

    return trained_models, results
```

3. Add predictions from ALL models

```python
def add_predictions(df, trained_models):
    X = df[["MA20", "MA50", "Volatility"]]

    for name, model in trained_models.items():
        df[f"Pred_{name}"] = model.predict(X)

    return df
```

4. FINAL SIGNAL (MAJORITY VOTING 🔥)

```python
def create_final_signal(df, model_names):
    pred_cols = [f"Pred_{name}" for name in model_names]

    # Majority vote
    df["Final_Prediction"] = df[pred_cols].mode(axis=1)[0]

    df["Signal"] = df["Final_Prediction"].map({
        1: "Bullish",
        0: "Bearish"
    })

    return df
```

ADD ADVANCED FEATURES (OHLC + OUTLIERS)

```python
def prepare_data(df):
    
    df = df.copy()

    # Returns
    df['Return'] = df['Close'].pct_change()

    # Moving averages
    df['MA20'] = df['Close'].rolling(20).mean()
    df['MA50'] = df['Close'].rolling(50).mean()

    # Volatility
    df['Volatility'] = df['Return'].rolling(20).std()

    # Target (1 = up, 0 = down)
    df['Target'] = (df['Close'].shift(-1) > df['Close']).astype(int)

    # Outliers (Z-score)
    df['Z'] = (df['Close'] - df['Close'].mean()) / df['Close'].std()
    df['Outlier'] = (df['Z'].abs() > 2)

    # Previous close
    df['Prev_Close'] = df['Close'].shift(1)

    df.dropna(inplace=True)

    return df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

def train_models(df, models):

    feature_cols = ['Return', 'MA20', 'MA50', 'Volatility']

    X = df[feature_cols]
    y = df['Target']

    X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False)

    trained_models = {}
    results = {}

    for name, model in models.items():
        model.fit(X_train, y_train)
        preds = model.predict(X_test)

        acc = accuracy_score(y_test, preds)

        trained_models[name] = model
        results[name] = acc

    return trained_models, results
```

## Importance of Stop Loss and Target in Stock Trading Systems

In any stock trading or investment strategy, defining Stop Loss and Target levels is essential for effective risk management and decision-making. These two components help transform a prediction system into a practical and disciplined trading framework.

##  Stop Loss – Controlling Risk

##### A stop loss is a predefined price level at which a position is automatically exited to prevent further losses. Financial markets are inherently uncertain, and even the most accurate machine learning models cannot guarantee 100% correct predictions. Therefore, stop loss acts as a safety mechanism.

### In this project, stop loss ensures that:

1. Losses are limited and controlled
2. Emotional decision-making is reduced
3. Capital is preserved for future trades
4. The system remains robust even during incorrect predictions

For example, if a stock is predicted as “Buy” but the market moves downward, the stop loss prevents excessive loss by exiting the trade early.

## Target – Securing Profits

##### A target price represents the expected level at which profits should be booked. It defines a clear exit strategy when the market moves in the predicted direction.

### In this system, target levels help:

1. Lock in profits systematically
2. Avoid greed-driven decisions (holding too long)
3. Maintain consistency in returns
4. Convert predictions into actionable outcomes

For instance, if a stock shows a bullish signal, the target ensures that profit is captured once the expected price movement is achieved.

### Risk-Reward Balance

. The combination of stop loss and target establishes a risk-reward ratio, which is a key principle in financial trading. A good trading strategy ensures that:

1. Potential profit (target) is greater than potential loss (stop loss)
2. Trades are taken only when the risk is justified by expected returns

## Role in AI-Based System

##### In an AI-Powered Automated Stock Report Generation System, predictions alone are not sufficient. The system must provide actionable insights. By integrating stop loss and target levels:

1. The model output becomes practically usable for traders
2. It simulates real-world trading conditions
3. It enhances the credibility and reliability of the system


## Conclusion

Stop loss and target levels are critical components that bridge the gap between theoretical prediction and real-world trading execution. They ensure disciplined trading, reduce financial risk, and make the AI system more aligned with professional financial practices. Without these mechanisms, even accurate predictions can lead to poor trading outcomes due to lack of proper exit strategies.
